# Problem 3

```python
import numpy as np
import time

A = np.random.normal(0, 1, (5000, 5000))
b = np.random.normal(0, 1, (5000, 1))
```

```python
def sol1(A, b):
    s = time.time()
    x = np.linalg.solve(A, b)
    e = time.time()
    return e-s

sol1(A, b)
```

```python
def sol2(A, b):
    s = time.time()
    A_inv = np.linalg.inv(A)
    x = np.matmul(A_inv, b)
    e = time.time()
    return e-s    
```

```python
sol2(A, b)
```

using `np.solve` is roughly 2 times faster in terms of running time.

# Problem 4

```python
import numpy as np

A = np.random.normal(0, 1, (100, 100))
u, s, vh = np.linalg.svd(A, full_matrices=True)
s_dagger = 1/s
```

```python
np.allclose(np.transpose(vh).dot(np.diag(s_dagger, k=0)).dot(np.transpose(u)), np.linalg.pinv(A))
```

# Problem 10

```python
# (a)
import numpy as np

N = [10, 100, 1000]

for n in N:
    print(n)
    A = np.random.normal(0, 1, (n,n))
    print(1/n * A.dot(np.transpose(A)))
```

It gets close to identity matrix $I$ as n goes to large

```python
# (b)
n=10
M = [10, 100, 1000]

for m in M:
    print(m)
    samples = []
    for _ in range(m):
        samples.append(np.random.normal(0, 1, (n,n)))
    print(1/m*sum([A.dot(A) for A in samples]))
```

It gets close to identity matrix $I$ as n goes to large
